### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create a 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle('2x2 Subplot Layout with Different Chart Types', fontsize=16)

# --- Subplot 1: Line Chart ---
x_line = np.arange(0, 10, 0.1)
y_line = np.sin(x_line)
axs[0, 0].plot(x_line, y_line, color='blue')
axs[0, 0].set_title('Line Chart (Sine Wave)')
axs[0, 0].set_xlabel('X-axis')
axs[0, 0].set_ylabel('Y-axis')
axs[0, 0].grid(True)

# --- Subplot 2: Bar Chart ---
categories_bar = ['Cat A', 'Cat B', 'Cat C', 'Cat D']
values_bar = [10, 15, 7, 12]
axs[0, 1].bar(categories_bar, values_bar, color=['skyblue', 'lightcoral', 'lightgreen', 'gold'])
axs[0, 1].set_title('Bar Chart (Categories)')
axs[0, 1].set_xlabel('Category')
axs[0, 1].set_ylabel('Value')
axs[0, 1].grid(axis='y', linestyle='--', alpha=0.7)

# --- Subplot 3: Scatter Plot ---
x_scatter = np.random.rand(50) * 10
y_scatter = np.random.rand(50) * 10
axs[1, 0].scatter(x_scatter, y_scatter, color='purple', alpha=0.7)
axs[1, 0].set_title('Scatter Plot (Random Data)')
axs[1, 0].set_xlabel('X-value')
axs[1, 0].set_ylabel('Y-value')
axs[1, 0].grid(True)

# --- Subplot 4: Pie Chart ---
labels_pie = ['Task 1', 'Task 2', 'Task 3', 'Task 4']
sizes_pie = [30, 20, 25, 25]
colors_pie = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']
axs[1, 1].pie(sizes_pie, labels=labels_pie, autopct='%1.1f%%', colors=colors_pie, startangle=90)
axs[1, 1].set_title('Pie Chart (Task Distribution)')
axs[1, 1].axis('equal') # Equal aspect ratio ensures that pie is drawn as a circle.

# Adjust layout to prevent overlapping titles/labels
plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Adjust rect to make space for suptitle

# Display the overall plot
plt.show()
```
