### Exercise 1: Line Plot Customization

```python
# TASK: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 1. Create x values ranging from 0 to 10
x = np.arange(0, 11)

# 2. Calculate y values as the square of x
y = x**2

# 3. Customize the plot by adding a title, labels for both axes, and a grid
plt.figure(figsize=(10, 6)) # Optional: set figure size for better readability
plt.plot(x, y, marker='o', linestyle='-') # marker and linestyle are optional but good for clarity
plt.title('Line Plot of y = x^2')
plt.xlabel('X-axis')
plt.ylabel('Y-axis (x^2)')
plt.grid(True) # Add a grid for better readability

# Display the plot
plt.show()
```
