### Exercise 6: Histogram Customization

```python
# TASK: Create a histogram for the following data: data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5].
# Customize the histogram with a title, labels for the x-axis, and a specific color for the bars.
```

```python
import matplotlib.pyplot as plt
import numpy as np
# Data for the histogram
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5]

# Create the histogram
plt.figure(figsize=(8, 6)) # Optional: set figure size
plt.hist(data, bins=range(min(data), max(data) + 2), rwidth=0.8, color='teal', edgecolor='black')

# Customize the histogram
plt.title('Histogram of Sample Data')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Set x-axis ticks to represent integer values clearly
plt.xticks(np.arange(min(data), max(data) + 1, 1))

# Add a grid for better readability (optional)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the plot
plt.show()
```
